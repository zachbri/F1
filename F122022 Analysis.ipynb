You may have noticed *DNF* and *DNS* in drivers' time, as well as *NC* in drivers' position. But what exactly does this mean?
* NC is defined as Not Classified. A driver is labeled as "Not Classified" in the race results if he completes **less than 90%** of the winner's race distance.
* DNF stands for "Did Not Finish." This is used to indicate that a driver has failed to finish a race due to various reasons.
* DNS on the other hand means "Did Not Start." This classification is used when a driver is unable to participate in the race at all. 

Various factors could cause a driver to be classified as a DNS or DNF. Common reasons for a **DNS** could include technical issues with a car before the race, a driver being unable to participate due to illness or injuries acquired during Free Practice or Qualifying, or a team deciding not to participate in the Grand Prix for strategic or logistical reasons. The same may apply for **DNF**, in which drivers could encounter mechanical failures and accidents that hindered them from finishing the race.

### DNF Analysis

```python
%%sql
SELECT 
    driver, 
    team,
    COUNT(CASE WHEN time = 'DNF' THEN 1 END) AS dnf_count,
    SUM(points) as total_points
FROM (
    SELECT driver, team, points, time FROM season_race_results
    UNION ALL
    SELECT driver, team, points, NULL AS time FROM sprint_results
) AS combined_results
GROUP BY driver, team
ORDER BY total_points DESC;
```

The table above shows the total number of DNFs each driver has alongside the total number of points they have accumulated during the season. From here, we can make a few analyses through the data:
* **Max Verstappen** was able to recover and dominate the 2022 season despite having 2 DNFs.
* **Charles Leclerc** and **Sergio Perez** were able to exhibit consistency in scoring points despite facing 3 DNFs each, which manifests their ability to capitilize on races that they have successfully completed. 
* Despite having 6 DNFs, **Carlos Sainz** displayed strong performance by finishing 6th overall in the season. This highlights the consistent challenges Ferrari faces in reliability and performance.
* There are potential areas for car development, team strategy development, and driver adaptation for drivers such as **Yuki Tsunoda, Guanyu Zhou, Alexander Albon**, and **Kevin Magnussen** as they encountered the most DNFs while accumulating the least points the entire season.

```python
%%sql
SELECT 
    team,
    COUNT(CASE WHEN time = 'DNF' THEN 1 END) AS dnf_count,
    SUM(points) as total_points
FROM (
    SELECT team, points, time FROM season_race_results
    UNION ALL
    SELECT team, points, NULL AS time FROM sprint_results
) AS combined_results
GROUP BY team
ORDER BY total_points DESC;
```

From a team perspective, **Red Bull** stood out by having one of the lowest numbers of DNFs (5) while achieving the highest total points, showcasing the team's strong overall performance and reliability. On the other hand, **Ferrari** faced a higher number of DNFs (9) but still managed to secure 2nd place in total team points, indicating the team's competitive performance in races they were able to complete. The bottom 5 teams had high numbers of DNFs and the lowest total point counts, illustrating the challenges these teams faced throughout the season.

### Qualifying Race
A standard Formula 1 grand prix weekend, excluding Sprint races *(which we'll be discussing later on)* typically comprises three segments: Free Practice, Qualifying Race, and the Grand Prix race itself. A qualifying session usually takes place the day before the actual race, commonly on Saturday. The format of the Qualifying session are as follows:
1. **Q1 (First Qualifying)**
* All participating cars are on the track
* The Q1 session lasts for 18 minutes long, followed by a 7 minute break
* The 5 cars who have produced the slowest lap time are eliminated and will start 16th to 20th on the grid depending on their lap time
2. **Q2 (Second Qualifying)**
* All remaining cars participate in Q2
* Q2 sessions last for 15 minutes, followed by an 8 minute break
* The 5 slowest cars will start 11th to 15th on the grid
3. **Q3 (Third Qualifying)**
* All remaining cars will compete for the first (or *pole*) position on the starting grid
* Q3 sessions last for 12 minutes long
* Top 10 grid positions are determined based on drivers' lap times


```python
%%sql
SELECT driver, COUNT(*) AS pole_positions
FROM quali_results
WHERE position = 1
GROUP BY driver
ORDER BY pole_positions DESC;
```

Claiming pole position in Formula 1 is often seen as a critical step towards victory, offering a clear path ahead and an early advantage in the race. ### Sprint Weekend
Sprint races are a recent addition to the racing format in Formula 1. Introduced in 2021, Sprint are shorter races held over a fixed distance, typically around **one-third** of the full Grand Prix distance. The format for a Grand Prix changes on weekends with sprint races
* **Friday:** There is only 1 practice session, followed by the standard qualifying session format. This session establishes the grid for Sunday's grand prix and does not affect the sprint qualifying or sprint race on Saturday.
* **Saturday:** Sprint qualifying takes place in the morning, following the structure of regular qualifying but with shorter time intervals for Q1 (12 minutes), Q2 (10 minutes), and Q3 (8 minutes). This is then followed by the sprint race, which awards the winner with 8 points, with points decreasing incrementally to 1 for the driver finishing eigth.
* **Sunday:** The grand prix race proceeds as usual

## 2022 Constructor Standings
A Constructor Championship, once known as the International Cup of for Formula One Manufacturers, is the highest sum of total points combined by each teams drivers at the end of the season in order to recognize the overall success of an entire team. This award recognizes the collective dedication of engineers, strategists, and drivers who bring their distinct expertise to achieve every teams goal. But beyond the numbers, the Constructor's Championship is a showcase of the sport's essenceprecision, innovation, and teamwork. 


```python
%%sql
SELECT team, SUM(points) as total_points
FROM (
    SELECT team, points FROM season_race_results
    UNION ALL
    SELECT team, points FROM sprint_results
) AS total_points
GROUP BY team
ORDER BY total_points DESC;
```

With **Verstappen's** exceptional dominance, **Red Bull Racing** secured a substantial lead in the **Formula 1 World Constructors' Championship**, outpacing **Ferrari** and **Mercedes** by over **200 points**.  For Mercedes fans, this sadly marked the end of **Mercedes'** impressive 7-year streak as Constructors' Champions.
